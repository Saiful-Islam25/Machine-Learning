```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Load dataset
df = pd.read_excel("house.xlsx")
```

```python
# Drop ID column
df = df.drop(columns=['Id'])
```

```python
# Separate target variable
y = df['SalePrice']
X = df.drop(columns=['SalePrice'])
```

```python
# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(exclude=['object']).columns
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
# Drop categorical columns with too many missing values
threshold = 0.5  # drop columns with >50% missing
missing_ratios = X[categorical_cols].isnull().mean()
cols_to_drop = missing_ratios[missing_ratios > threshold].index
X = X.drop(columns=cols_to_drop)
categorical_cols = [col for col in categorical_cols if col not in cols_to_drop]
```

```python
# Fill missing categorical values with 'None'
X[categorical_cols] = X[categorical_cols].fillna('None')
```

```python
X[categorical_cols]
```

```python
# Fill missing numerical values with median
X[numerical_cols] = X[numerical_cols].fillna(X[numerical_cols].median())
```

```python
X[numerical_cols]
```

```python
# Ordinal encoding for quality-related columns
ordinal_mappings = {
    "Ex": 5,
    "Gd": 4,
    "TA": 3,
    "Fa": 2,
    "Po": 1,
    "None": 0
}

ordinal_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC',
                'KitchenQual', 'FireplaceQu', 'GarageQual', 'GarageCond',
                'PoolQC']

for col in ordinal_cols:
    if col in X.columns:
        X[col] = X[col].map(ordinal_mappings).fillna(0).astype(int)
```

```python
# One-hot encode remaining categorical features
X = pd.get_dummies(X, columns=[col for col in categorical_cols if col not in ordinal_cols], drop_first=True)
```
