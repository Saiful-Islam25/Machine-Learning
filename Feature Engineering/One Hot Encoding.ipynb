```python
import pandas as pd
```

```python
data={'Category': ['A','B','C','A','B','C','A','B','C']}
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
encoded_df=pd.get_dummies(df,columns=['Category'])
encoded_df
```

```python
encoded_df=pd.get_dummies(df,columns=['Category'],prefix='Dummy')
encoded_df
```

```python
encoded_df=pd.get_dummies(df,columns=['Category'],prefix='Dummy',drop_first=True)
encoded_df
```
